```python
import json
import time
from deep_translator import GoogleTranslator

# Caminhos
input_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\teste\RJH006_subj_obj_transcription_words.json"
output_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\teste\transcription_test_english.json"

# Tradutor: de chinês para inglês
translator = GoogleTranslator(source='zh-TW', target='en')

# Carrega o JSON original
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# Traduz apenas os primeiros 2 segmentos (ajuste aqui se quiser)
subset = data["sentences"][:100]

for i, segment in enumerate(subset):
    try:
        segment["text_eng"] = translator.translate(segment["text"])
        print(f"✅ Segmento {i+1}: {segment['text']} -> {segment['text_eng']}")
        
        for word in segment["words"]:
            word["text_eng"] = translator.translate(word["text"])
            time.sleep(0.2)
    except Exception as e:
        print(f"⚠️ Erro ao traduzir segmento {i+1}: {e}")
    time.sleep(1)

# Cria um novo JSON com apenas os segmentos traduzidos
output_data = {"segments": subset}

# Salva o resultado
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(output_data, f, ensure_ascii=False, indent=2)

print(f"\n✅ Tradução de teste concluída. Arquivo salvo em:\n{output_path}")
```
