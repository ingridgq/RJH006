```python
import json
import re
import os 

#Paths
folder_base = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj"
experiment_name = "RJH006_subj_obj"
#json_path = f"{folder_base}/{experiment_name}_transcription.json"
json_path = f"{folder_base}/transcription_english_NEW.json"

#Names to find
names = [
    {"text": "朱茵", "text_eng": "Zhu Yin"},
    {"text": "焦恩俊", "text_eng": "Jiao Enjun"},
    {"text": "周星驰", "text_eng": "Stephen Chow"},
    {"text": "金映娟", "text_eng": "Jin Yingjuan"},
    {"text": "张怡宁", "text_eng": "Zhang Yining"},
]

#Original transcription
with open(json_path, 'r', encoding='utf-8') as f: 
    data = json.load(f)

results = []
ids_list = []

for seg in data.get("segments", []):
    eng_sentence = seg.get("text_eng", [])

    words = seg.get("words", [])
    full_text = "".join([w["text"] for w in words])

#---------------------------------------------------------------------

    new_words = []

    for x, name in enumerate(names): 
        n = name["text"]
        eng = name["text_eng"] 

        #Verify if name is inside the segment
        if n in full_text: 
            text_concat = ""
            for i in range(len(words)):
                text_concat += words[i]["text"]

                if text_concat.endswith(n): 
                    start_idx = i - len(n) + 1
                    end_idx = i
                    start_time = words[start_idx]["start"]
                    end_time = words[end_idx]["end"]

                    sentence_start = words[0]["start"] 
                    sentence_end = words[-1]["end"]

                    speaker_id = words[0]["speaker_id"] 

                    results.append({
                        "chin_name": n,
                        "eng_name": eng,
                        "start": start_time,
                        "end": end_time,
                        "chin_sentence": full_text,
                        "eng_sentence": eng_sentence,
                        "sentence_start": sentence_start,
                        "sentence_end": sentence_end,
                        "speaker_id" : speaker_id
                        })
                    break 

                #testing ------------------------------------------------------
            current_new_word = []
            start_time = None

            for w in words:
                t = w["text"]
                s = w["start"]
                e = w["end"]

                #Separating words
                if re.match(r"[，。！？、,.!? \s]+", t):
                    if current_new_word:
                        new_word_text = "".join([cw["text"] for cw in current_new_word])
                        new_word_start = current_new_word[0]["start"]
                        new_word_end = current_new_word[-1]["end"]
                        new_words.append({
                            "text": new_word_text,
                            "start": new_word_start,
                            "end": new_word_end
                        })
                        current_new_word = []
                    #The potuation isn't saved
                else:
                    #Incorporates the caracter as part of the word
                    current_new_word.append(w)

            #Saves the end
            if current_new_word:
                new_word_text = "".join([cw["text"] for cw in current_new_word])
                new_word_start = current_new_word[0]["start"]
                new_word_end = current_new_word[-1]["end"]
                new_words.append({
                    "text": new_word_text,
                    "start": new_word_start,
                    "end": new_word_end
                })


            splited_words = [nw["text"] for nw in new_words]
            for sw in splited_words: 
                if n in sw: 
                    id_n = sw.find(n)

                    ids_list.append({"sentence": full_text,
                        "name": n,
                         "id_n": id_n
                        })
                    
print(ids_list)
        
            #if ids_list["name"].item

                

            
```
