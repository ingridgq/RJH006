```python
import os
import json
import re

def transcribe_words(folder_base, experiment_name):

    json_path = f'{folder_base}/{experiment_name}_transcription.json'

    #Reading JSON
    with open(json_path, 'r', encoding='utf-8') as f_json:
        json_data = json.load(f_json)

    new_words = []

    for seg in json_data["segments"]:
        words = seg.get("words", [])
        if not words:
            continue

        current_new_word = []
        start_time = None

        for w in words:
            t = w["text"]
            s = w["start"]
            e = w["end"]

            #Separating words
            if re.match(r"[，。！？、,.!? \s]+", t):
                if current_new_word:
                    new_word_text = "".join([cw["text"] for cw in current_new_word])
                    new_word_start = current_new_word[0]["start"]
                    new_word_end = current_new_word[-1]["end"]
                    new_words.append({
                        "text": new_word_text,
                        "start": new_word_start,
                        "end": new_word_end
                    })
                    current_new_word = []
                #The potuation isn't saved
            else:
                #Incorporates the caracter as part of the word
                current_new_word.append(w)

        #Saves the end
        if current_new_word:
            new_word_text = "".join([cw["text"] for cw in current_new_word])
            new_word_start = current_new_word[0]["start"]
            new_word_end = current_new_word[-1]["end"]
            new_words.append({
                "text": new_word_text,
                "start": new_word_start,
                "end": new_word_end
            })

    #Saving
    with open(f'{folder_base}/{experiment_name}_transcription_words.json', 'w', encoding='utf-8') as f_out:
        json.dump({"sentences":  new_words}, f_out, ensure_ascii=False, indent=2)

    print(f"✅ New words file saved at: f'{folder_base}/{experiment_name}_transcription_words.json")
```

```python
folder_base = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\teste"
audio_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\teste\RJH006_subj_obj_audio.wav"
experiment_name = "RJH006_subj_obj"
transcribe_words(folder_base, experiment_name)
```
