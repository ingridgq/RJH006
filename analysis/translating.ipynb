```python
import time
import json
from deep_translator import GoogleTranslator

input_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\RJH006_subj_obj_transcription.json"
output_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\transcription_english.json"

translator = GoogleTranslator(source='zh-TW', target='en')

with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

def safe_translate(text, retries=5):
    """Traduz com tentativas e pausas em caso de falha."""
    for attempt in range(retries):
        try:
            return translator.translate(text)
        except Exception as e:
            print(f"‚ö†Ô∏è Erro na tradu√ß√£o ({e}). Tentando novamente ({attempt+1}/{retries})...")
            time.sleep(5 * (attempt + 1))  # espera cada vez mais tempo
    print("‚ùå Falhou ap√≥s v√°rias tentativas.")
    return text  # retorna o texto original para n√£o perder dados

for i, segment in enumerate(data["segments"]):
    segment["text_eng"] = safe_translate(segment["text"])
    time.sleep(0.5)

    for word in segment["words"]:
        word["text_eng"] = safe_translate(word["text"])
        time.sleep(0.2)

    print(f"‚úÖ Segmento {i+1}/{len(data['segments'])} traduzido.")

    # üíæ salva progresso parcial a cada 10 segmentos
    if (i + 1) % 10 == 0:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)
        print(f"üíæ Progresso salvo ({i+1} segmentos).")

# salva tudo no final
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"‚úÖ Tradu√ß√£o conclu√≠da e salva em: {output_path}")
```
