```python
import json
import csv
import pandas as pd

#Paths
json_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\RJH006_subj_obj_names_timestampsNEW.json"
csv_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\RJH006_subj_obj_names_timestamps_CSV.csv"
excel_path = r"C:\Users\ingri\Documents\IMIM\experiments\RJH006_subj_obj\RJH006_subj_obj_names_timestamps_XLSX.xlsx"

#Opening json
with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

#criating and writing csv
with open(csv_path, "w", newline="", encoding="utf-8-sig") as csvfile:
    writer = csv.writer(csvfile)

    #headers
    writer.writerow(["chinese_name", "english_name", "start_time", "end_time", "chinese_sentence", "english_sentence", "sentence_start", "sentence_end"])

    #iterating
    for item in data:
        chinese = item.get("chin_name", "")
        english = item.get("eng_name", "")
        start = item.get("start", "")
        end = item.get("end", "")
        chin_sentence = item.get("chin_sentence", "")
        eng_sentence = item.get("eng_sentence", "")
        sentence_start = item.get("sentence_start", "")
        sentence_end = item.get("sentence_end", "")

        writer.writerow([chinese, english, start, end, chin_sentence, eng_sentence, sentence_start, sentence_end])

print(f'.csv succesfully criated: {csv_path}') 

# Salva como Excel (.xlsx)
df = pd.read_csv(csv_path)
df.to_excel(excel_path, index=False)

print(f'.xlsx succesfully criated: {excel_path}')
```
